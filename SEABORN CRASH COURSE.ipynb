**Visualizing Statistical relationships**

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationship depends on other variable. Visualization can be a core components of this process because when data are visualized properly, the human visual system can see the treds and patterns that indicate a relationship

```python
##Importing liabraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##setting style
sns.set(style='darkgrid')
```

```python
#loading dataset
tips = sns.load_dataset('tips')
tips.head()
```

**1. Numerical Data Plotting**

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
```

```python
#changing color
sns.relplot(x='total_bill',y='tip',data=tips,hue='size',palette='ch:r=-0.2,l=0.75')
```

```python
x = randn(100)
```

```python
sns.distplot(x)
```

```python
sns.distplot(x,kde=False,rug=True,bins=30)
```

```python
sns.kdeplot(x,shade=True,cbar=True)
```

```python
#Bivariate 
tips.head()
```

```python
x =tips['total_bill']
y=tips['tip']
```

```python
sns.jointplot(x=x,y=y,kind='hex')
```

```python
sns.jointplot(x=x,y=y,kind='kde')
```

```python
g = sns.jointplot(x,y,kind='kde',color='m')
g.plot_joint(plt.scatter,c='w',s=30,linewidth=1,marker='+')
g.ax_joint.collections[0].set_alpha(0)
```

4. Linear Regression and Relationship

regplot()

lmplot()

```python
tips.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='size',y='tip',data=tips,x_jitter=0.05)
```

```python
sns.lmplot(x='size',y='tip',data=tips,x_estimator=np.mean)
```

```python
#loading dataset
data = sns.load_dataset('anscombe')
```

```python
sns.lmplot(x='x',y='y',data=data.query("dataset=='II'"))
```

```python
#Removing outlier
```

```python
sns.lmplot(x='x',y='y',data=data.query("dataset=='I'"),robust=True)
```
